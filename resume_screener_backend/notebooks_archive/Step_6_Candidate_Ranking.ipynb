```python
import pandas as pd

INPUT_FILE = "/content/drive/MyDrive/resume_project/resume_jd_ats_score_v2_final.csv"
OUTPUT_FILE = "/content/drive/MyDrive/resume_project/resume_jd_ranked_final.csv"

df = pd.read_csv(INPUT_FILE)

df = df.sort_values(
    by=["job_position_name", "ats_score", "cosine_similarity"],
    ascending=[True, False, False]
)

df["candidate_rank"] = (
    df.groupby("job_position_name").cumcount() + 1
)

df.to_csv(OUTPUT_FILE, index=False)

print("Saved:", OUTPUT_FILE)

job = df["job_position_name"].iloc[0]

print(f"\nTop 5 candidates for '{job}':")

print(
    df[df["job_position_name"] == job]
    .head(5)[
        ["candidate_rank", "ats_score", "cosine_similarity", "matched_score"]
    ]
)
```
